```python
# importe do database

# perguntas para responder

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('data.csv')

df
```

```python

# Agrupar por gênero e calcular a média de uso diário de celular
df_horas_genero = df.groupby("Gender")["Avg_Daily_Usage_Hours"].mean().reset_index()

# Encontrar o gênero com o maior uso diário
faixa_genero_mais_uso = df_horas_genero.loc[df_horas_genero["Avg_Daily_Usage_Hours"].idxmax()]

# Plotar o gráfico de barras com cores diferentes
sns.barplot(data=df_horas_genero, x="Gender", y="Avg_Daily_Usage_Hours", palette="Blues")

# Títulos e rótulos
plt.title("Total Hours Day by Gender", fontsize=14)
plt.xlabel("Gender", fontsize=12)
plt.ylabel("Total Hours Day", fontsize=12)

# Adicionar os valores nas barras
for index, value in enumerate(df_horas_genero["Avg_Daily_Usage_Hours"]):
    plt.text(index, value + 0.1, f'{value:.2f}', ha='center', va='bottom', fontsize=11)

# Mostrar o gráfico
plt.show()
```

  com isso podemos concluir que o gênero feminino
  tem uma média maior de horas totais diárias de
  uso de dispositivos eletrônicos.

```python

# Agrupar por idade e calcular a média de horas diárias de uso de celular
df_uso_por_idade = df.groupby("Age")["Avg_Daily_Usage_Hours"].mean().reset_index()

# Identificar a faixa etária com o maior uso médio
faixa_etaria_mais_uso = df_uso_por_idade.loc[df_uso_por_idade["Avg_Daily_Usage_Hours"].idxmax()]


# Plotar o gráfico de barras
sns.barplot(data=df_uso_por_idade, x="Age", y="Avg_Daily_Usage_Hours", palette="Blues")

# Títulos e rótulos
plt.title("Uso médio de celular por idade", fontsize=14)
plt.xlabel("Idade", fontsize=12)
plt.ylabel("Horas médias de uso diário", fontsize=12)

# Mostrar o gráfico
plt.show()


```

A análise mostra que pessoas com 18 anos ou mais tendem a usar mais o celular. A média de horas de uso diário foi calculada por faixa etária, revelando que indivíduos dessa faixa etária têm um tempo de uso mais alto em comparação com outros grupos. Isso pode refletir o aumento do uso de celulares para atividades como trabalho, redes sociais e comunicação.

```python
# Agrupar por apps e calcular a média de app mais usado no dia
df_app_mais_usado = df.groupby("Most_Used_Platform")["Avg_Daily_Usage_Hours"].mean().reset_index()

# Identificar o app de mais uso
faixa_app_mais_usado = df_app_mais_usado.loc[df_app_mais_usado["Avg_Daily_Usage_Hours"].idxmax()]

# Plotar o gráfico de barras
sns.barplot(
    data=df_horas_genero,
    x="Gender",
    y="Avg_Daily_Usage_Hours",
    hue="Gender",
    palette="Blues",
    legend=False
)
# Títulos e rótulos
plt.title("Uso Médio de Celular por Aplicativo", fontsize=14)
plt.xlabel("App", fontsize=12)
plt.ylabel("Horas médias de uso diário", fontsize=12)

# Rotacionar os rótulos do eixo X para evitar sobreposição
plt.xticks(rotation=45, ha='right')

# Mostrar o gráfico
plt.show()
```

O gráfico mostra o uso médio diário de diferentes aplicativos entre os participantes da pesquisa. O WhatsApp aparece como o aplicativo mais utilizado, com uma média de uso mais alta, o que pode refletir seu papel importante na comunicação diária e na manutenção de amizades e interações no trabalho. Em segundo lugar, temos o TikTok, que se destaca como a rede social mais popular entre os jovens, com um crescimento significativo no engajamento, principalmente para entretenimento e consumo de conteúdo."

```python

# Agrupar por idade e calcurar vicio
df_vicio_por_idade = df.groupby("Age")["Addicted_Score"].mean().reset_index()

# Identificar a faixa com mais vicio me celular
faixa_vicio_maior = df_uso_por_idade.loc[df_vicio_por_idade["Addicted_Score"].idxmax()]


# Plotar o gráfico de barras
sns.barplot(
    data=df_vicio_por_idade,
    x="Age",
    y="Addicted_Score",
    hue="Age",
    palette="Blues",
    legend=False
)
# Títulos e rótulos
plt.title("Vicio em celular por idade", fontsize=14)
plt.xlabel("Idade", fontsize=12)
plt.ylabel("vicio", fontsize=12)

# Mostrar o gráfico
plt.show()

```

Pessoas com 18 anos apresentaram o maior tempo médio de uso diário do celular, indicando um possível maior nível de dependência ou hábito de uso intensivo entre os jovens dessa faixa etária.


```python

# Agrupar por idade e calcurar vicio
df_vicio_por_app = df.groupby("Most_Used_Platform")["Addicted_Score"].mean().reset_index()

# Identificar a faixa com mais vicio me celular
faixa_vicio_maior = df_vicio_por_app.loc[df_vicio_por_app["Addicted_Score"].idxmax()]


# Plotar o gráfico de barras
sns.barplot(
    data=df_vicio_por_app,
    x="Most_Used_Platform",
    y="Addicted_Score",
    hue="Most_Used_Platform",
    palette="Blues",
    legend=False
)

# Títulos e rótulos
plt.title("Vicio em celular por app", fontsize=14)
plt.xlabel("App", fontsize=12)
plt.ylabel("Vicio", fontsize=12)

# Rotacionar os rótulos do eixo X para evitar sobreposição
plt.xticks(rotation=45, ha='right')

# Mostrar o gráfico
plt.show()
```

Os aplicativos mais associados ao uso excessivo do celular foram TikTok, Snapchat e WhatsApp. O TikTok e o Snapchat se destacam como os mais utilizados para entretenimento, enquanto o WhatsApp, embora também muito usado, pode refletir tanto relações sociais quanto obrigações de trabalho ou estudo.

```python

# Agrupar por idade e calcurar vicio
df_confiltos_por_idade = df.groupby("Age")["Conflicts_Over_Social_Media"].mean().reset_index()

# Identificar a faixa com mais vicio me celular
faixa_confiltos_por_idade = df_confiltos_por_idade.loc[df_confiltos_por_idade["Conflicts_Over_Social_Media"].idxmax()]


# Plotar o gráfico de barras
sns.barplot(
    data=df_confiltos_por_idade,
    x="Age",
    y="Conflicts_Over_Social_Media",
    hue="Age",
    legend=False
# Títulos e rótulos
plt.title("Confiltos socias por Idade", fontsize=14)
plt.xlabel("Idade", fontsize=12)
plt.ylabel("Confiltos", fontsize=12)

# Mostrar o gráfico
plt.show()
```

```python
# Calcular a porcentagem de pessoas acima de 5h por idade
df_porcentagem = (
    df.assign(Acima_5h = df["Avg_Daily_Usage_Hours"] > 5)
      .groupby("Age")["Acima_5h"]
      .mean()
      .mul(100)
      .reset_index(name="Porcentagem_Acima_5h")
)

# Gráfico de barras
sns.barplot(
    data=df_porcentagem,
    x="Age",
    y="Porcentagem_Acima_5h",
    hue="Age",
    palette="coolwarm",
    legend=False
)
plt.title("Porcentagem de Usuários com Uso Acima de 5h por Idade")
plt.xlabel("Idade")
plt.ylabel("Porcentagem (%)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
