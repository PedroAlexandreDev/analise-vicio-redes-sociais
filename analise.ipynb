```python
# importe do database

# perguntas para responder

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('data.csv')

df
```

```python

# Agrupar por gÃªnero e calcular a mÃ©dia de uso diÃ¡rio de celular
df_horas_genero = df.groupby("Gender")["Avg_Daily_Usage_Hours"].mean().reset_index()

# Encontrar o gÃªnero com o maior uso diÃ¡rio
faixa_genero_mais_uso = df_horas_genero.loc[df_horas_genero["Avg_Daily_Usage_Hours"].idxmax()]

# Plotar o grÃ¡fico de barras com cores diferentes
sns.barplot(data=df_horas_genero, x="Gender", y="Avg_Daily_Usage_Hours", palette="Blues")

# TÃ­tulos e rÃ³tulos
plt.title("Total Hours Day by Gender", fontsize=14)
plt.xlabel("Gender", fontsize=12)
plt.ylabel("Total Hours Day", fontsize=12)

# Adicionar os valores nas barras
for index, value in enumerate(df_horas_genero["Avg_Daily_Usage_Hours"]):
    plt.text(index, value + 0.1, f'{value:.2f}', ha='center', va='bottom', fontsize=11)

# Mostrar o grÃ¡fico
plt.show()
```

  com isso podemos concluir que o gÃªnero feminino
  tem uma mÃ©dia maior de horas totais diÃ¡rias de
  uso de dispositivos eletrÃ´nicos.

```python

# Agrupar por idade e calcular a mÃ©dia de horas diÃ¡rias de uso de celular
df_uso_por_idade = df.groupby("Age")["Avg_Daily_Usage_Hours"].mean().reset_index()

# Identificar a faixa etÃ¡ria com o maior uso mÃ©dio
faixa_etaria_mais_uso = df_uso_por_idade.loc[df_uso_por_idade["Avg_Daily_Usage_Hours"].idxmax()]

# Exibir o resultado
print(f"A faixa etÃ¡ria com maior uso mÃ©dio de celular Ã©: {faixa_etaria_mais_uso['Age']} anos com {faixa_etaria_mais_uso['Avg_Daily_Usage_Hours']} horas por dia.")

# Plotar o grÃ¡fico de barras
sns.barplot(data=df_uso_por_idade, x="Age", y="Avg_Daily_Usage_Hours", palette="Blues")

# TÃ­tulos e rÃ³tulos
plt.title("Uso mÃ©dio de celular por idade", fontsize=14)
plt.xlabel("Idade", fontsize=12)
plt.ylabel("Horas mÃ©dias de uso diÃ¡rio", fontsize=12)

# Mostrar o grÃ¡fico
plt.show()


```

A anÃ¡lise mostra que pessoas com 18 anos ou mais tendem a usar mais o celular. A mÃ©dia de horas de uso diÃ¡rio foi calculada por faixa etÃ¡ria, revelando que indivÃ­duos dessa faixa etÃ¡ria tÃªm um tempo de uso mais alto em comparaÃ§Ã£o com outros grupos. Isso pode refletir o aumento do uso de celulares para atividades como trabalho, redes sociais e comunicaÃ§Ã£o.

```python
# Agrupar por apps e calcular a mÃ©dia de app mais usado no dia
df_app_mais_usado = df.groupby("Most_Used_Platform")["Avg_Daily_Usage_Hours"].mean().reset_index()

# Identificar o app de mais uso
faixa_app_mais_usado = df_app_mais_usado.loc[df_app_mais_usado["Avg_Daily_Usage_Hours"].idxmax()]

# Plotar o grÃ¡fico de barras
sns.barplot(data=df_app_mais_usado, x="Most_Used_Platform", y="Avg_Daily_Usage_Hours", palette="Blues")

# TÃ­tulos e rÃ³tulos
plt.title("Uso MÃ©dio de Celular por Aplicativo", fontsize=14)
plt.xlabel("App", fontsize=12)
plt.ylabel("Horas mÃ©dias de uso diÃ¡rio", fontsize=12)

# Rotacionar os rÃ³tulos do eixo X para evitar sobreposiÃ§Ã£o
plt.xticks(rotation=45, ha='right')

# Mostrar o grÃ¡fico
plt.show()
```

O grÃ¡fico mostra o uso mÃ©dio diÃ¡rio de diferentes aplicativos entre os participantes da pesquisa. O WhatsApp aparece como o aplicativo mais utilizado, com uma mÃ©dia de uso mais alta, o que pode refletir seu papel importante na comunicaÃ§Ã£o diÃ¡ria e na manutenÃ§Ã£o de amizades e interaÃ§Ãµes no trabalho. Em segundo lugar, temos o TikTok, que se destaca como a rede social mais popular entre os jovens, com um crescimento significativo no engajamento, principalmente para entretenimento e consumo de conteÃºdo."

```python

# Agrupar por idade e calcurar vicio
df_vicio_por_idade = df.groupby("Age")["Addicted_Score"].mean().reset_index()

# Identificar a faixa com mais vicio me celular
faixa_vicio_maior = df_uso_por_idade.loc[df_vicio_por_idade["Addicted_Score"].idxmax()]


# Plotar o grÃ¡fico de barras
sns.barplot(data=df_vicio_por_idade, x="Age", y="Addicted_Score", palette="Blues")

# TÃ­tulos e rÃ³tulos
plt.title("Vicio em celular por idade", fontsize=14)
plt.xlabel("Idade", fontsize=12)
plt.ylabel("vicio", fontsize=12)

# Mostrar o grÃ¡fico
plt.show()

```
